```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,100)
a = 3*t
b = 5*t-10
```

```python
fig, ax = plt.subplots()

plt.title('System of equations')
plt.xlabel('X axis')
plt.ylabel('Y axis')

ax.set_xlim([0,100])
ax.set_ylim([0,100])

ax.plot(t, a, c = 'red')
ax.plot(t, b, c = 'blue')

plt.axvline(x = 5)
plt.axhline(y = 10)
```
